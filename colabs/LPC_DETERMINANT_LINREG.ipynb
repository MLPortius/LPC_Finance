```python
#@title LINREG SETUP
grid_dataset = 1 #@param {type:"integer"}
depvar = "da" #@param ["da", "mape", "dis"]
missing_values = "drop" #@param ["mean", "drop","cero"]
feature_selection = True #@param {type:"boolean"}
select_percent = 50 #@param {type:"slider", min:0, max:100, step:5}
```

```python
#@title SELECT VARIABLES
SPREAD_BETA = True #@param {type:"boolean"}
MARKET_BETA = True #@param {type:"boolean"}
R_MSTD_ST = True #@param {type:"boolean"}
R_MSTD_LT = True #@param {type:"boolean"}
W_SIZE = True #@param {type:"boolean"}
P_LAGS = True #@param {type:"boolean"}
MARKETCAP_RMEAN = True #@param {type:"boolean"}
VOLUME_CV = True #@param {type:"boolean"}
PBPERSHARE_RMEAN = True #@param {type:"boolean"}
STREAK_HMEAN = True #@param {type:"boolean"}
STREAK_PROBA = True #@param {type:"boolean"}
MAX_STREAK = True #@param {type:"boolean"}
ADX30D_TREND_DAYS = True #@param {type:"boolean"}
ADX30D_TREND_PROBA = True #@param {type:"boolean"}
VIX_BETA = True #@param {type:"boolean"}
ANALYST_DISPERSION_RMEAN = True #@param {type:"boolean"}
ANALYST_NUMBER_RMEAN = True #@param {type:"boolean"}
PUTCALLRATIO_RMEAN = True #@param {type:"boolean"}
BIDASK_RMEAN = True #@param {type:"boolean"}
PRICE_CUSUM_BREAKS = True #@param {type:"boolean"}
HURST_RANDWALK_DAYS = True #@param {type:"boolean"}
HURST_RANDWALK_PROBA = True #@param {type:"boolean"}
HURST_MEANREV_DAYS = True #@param {type:"boolean"}
HURST_MEANREV_PROBA = True #@param {type:"boolean"}
HURST_TRENDING_DAYS = True #@param {type:"boolean"}
HURST_TRENDING_PROBA = True #@param {type:"boolean"}
INDUSTRIES  = True #@param {type:"boolean"}
```

```python
#@title PREPARAR ENTORNO
import os
import numpy as np
import pandas as pd
import compress_pickle as cpickle
import statsmodels.api as sm
from sklearn.feature_selection import mutual_info_regression

dset = 'set'+str(grid_dataset)
dpath = dset + '/'
inpath = 'output/determinants/'+dpath
depath = inpath + depvar + '/'
```

```python
#@title CARGAR Y PREPARAR DATOS
data = cpickle.load(inpath + 'det_main_data.lzma')

if depvar == 'da':
  y = data.loc[:,'OPT_DA']
  x = data.drop(['OPT_DA','OPT_MAPE','OPT_DIS'], axis=1)
  x.drop(['DIS_W_SIZE','DIS_P_LAGS','MAPE_W_SIZE','MAPE_P_LAGS'], axis=1, inplace=True)

elif depvar == 'mape':
  y = data.loc[:,'OPT_MAPE']
  x = data.drop(['OPT_DA','OPT_MAPE','OPT_DIS'], axis=1)
  x.drop(['DIS_W_SIZE','DIS_P_LAGS','DA_W_SIZE','DA_P_LAGS'], axis=1, inplace=True)

elif depvar == 'dis':
  y = data.loc[:,'OPT_DIS']
  x = data.drop(['OPT_DA','OPT_MAPE','OPT_DIS'], axis=1)
  x.drop(['DA_W_SIZE','DA_P_LAGS','MAPE_W_SIZE','MAPE_P_LAGS'], axis=1, inplace=True)

if SPREAD_BETA == False:
  x.drop(['SPREAD_BETA'], axis=1)

if MARKET_BETA == False:
  x.drop(['MARKET_BETA'], axis=1)

if R_MSTD_ST == False:
  x.drop(['R_MSTD_ST'], axis=1)

if R_MSTD_LT == False:
  x.drop(['R_MSTD_LT'], axis=1)

if W_SIZE == False:
  if depvar == 'da':
    x.drop(['DA_W_SIZE'], axis=1)
  elif depvar == 'mape':
    x.drop(['MAPE_W_SIZE'], axis=1)
  elif depvar == 'dis':
    x.drop(['DIS_W_SIZE'], axis=1)

if P_LAGS == False:
  if depvar == 'da':
    x.drop(['DA_P_LAGS'], axis=1)
  elif depvar == 'mape':
    x.drop(['MAPE_P_LAGS'], axis=1)
  elif depvar == 'dis':
    x.drop(['DIS_P_LAGS'], axis=1)

if MARKETCAP_RMEAN == False:
  x.drop(['MARKETCAP_RMEAN'], axis=1)

if VOLUME_CV == False:
  x.drop(['VOLUME_CV'], axis=1)

if PBPERSHARE_RMEAN == False:
  x.drop(['PBPERSHARE_RMEAN'], axis=1)

if STREAK_HMEAN == False:
  x.drop(['STREAK_HMEAN'], axis=1)

if STREAK_PROBA == False:
  x.drop(['STREAK_PROBA'], axis=1)

if MAX_STREAK == False:
  x.drop(['MAX_STREAK'], axis=1)

if ADX30D_TREND_DAYS == False:
  x.drop(['ADX30D_TREND_DAYS'], axis=1)

if ADX30D_TREND_PROBA == False:
  x.drop(['ADX30D_TREND_PROBA'], axis=1)

if VIX_BETA == False:
  x.drop(['VIX_BETA'], axis=1)

if ANALYST_DISPERSION_RMEAN == False:
  x.drop(['ANALYST_DISPERSION_RMEAN'], axis=1)

if ANALYST_NUMBER_RMEAN == False:
  x.drop(['ANALYST_NUMBER_RMEAN'], axis=1)

if PUTCALLRATIO_RMEAN == False:
  x.drop(['PUTCALLRATIO_RMEAN'], axis=1)

if BIDASK_RMEAN == False:
  x.drop(['BIDASK_RMEAN'], axis=1)

if PRICE_CUSUM_BREAKS == False:
  x.drop(['PRICE_CUSUM_BREAKS'], axis=1)

if HURST_RANDWALK_DAYS == False:
  x.drop(['HURST_RANDWALK_DAYS'], axis=1)

if HURST_RANDWALK_PROBA == False:
  x.drop(['HURST_RANDWALK_PROBA'], axis=1)

if HURST_MEANREV_DAYS == False:
  x.drop(['HURST_MEANREV_DAYS'], axis=1)

if HURST_MEANREV_PROBA == False:
  x.drop(['HURST_MEANREV_DAYS'], axis=1)

if HURST_TRENDING_DAYS == False:
  x.drop(['HURST_MEANREV_DAYS'], axis=1)

if HURST_TRENDING_PROBA == False:
  x.drop(['HURST_MEANREV_DAYS'], axis=1)

if INDUSTRIES == False:
  x = x.iloc[:,:-25]


if missing_values == 'mean':
  means = {}
  for c in list(x.columns):
    means[c] = x[c].mean()
  for c in list(x.columns):
    x[c] = x[c].fillna(means[c])

elif missing_values == 'drop':
  xy = pd.concat([x, y], axis=1)
  xy.dropna(axis=0, inplace=True)
  
  x = xy.iloc[:,:-1]
  y = xy.iloc[:,-1]
  
elif missing_values == 'cero':
  x.fillna(0, inplace=True)


if feature_selection:

  info = mutual_info_regression(x, y , random_state=14)
  info = pd.DataFrame(info, index=x.columns, columns=['INFO'])
  info.sort_values(['INFO'],ascending=False, inplace=True)

  n = len(list(x.columns))
  k = int(np.round(n * (select_percent/100), 0))

  selected = info.iloc[:k]
  selected_features = list(selected.index)

  xs = x.loc[:,selected_features]

else:
  xs = x.copy()
```

```python
#@title REGRESION LINEAL

xsc = sm.add_constant(xs)

model = sm.OLS(y, xsc)
results = model.fit()

print(results.summary())
```
